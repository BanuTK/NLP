```python
document=Document() # this will create a document object
document.add_heading('Document Title', 0)
p=document.add_paragraph('A plain paragrapgh having some')
p.add_run(' bold').bold=True
p.add_run(' and some ')
p.add_run('italic.').italic=True

document.save('NLPdemo.docx')
```

```python
document.add_heading('Heading, level1', level=1)
document.add_paragraph('Intense quote', style='Intense Quote')
document.add_paragraph(
        'first item in unordered list', style='List Bullet'
)
document.add_paragraph(
        'first item in ordered list', style='List Number'
)
document.add_picture('nlpdemo.png', width=Inches(1.25))
records=(
        (3,  '101', 'Spam') ,
        (7, '422', 'Eggs') ,
        (4, '631', 'Spam, spam, spam, eggs, and spam')
)
table=document.add_table(rows=1, cols=3)
hdr_cells=table.rows[0].cells
hdr_cells[0].text='Qty' #cell 0 heading will be Qty, cell means COLUMNS
hdr_cells[1].text='Id'  #column heading
hdr_cells[2].text='Desc'
for qty, id, desc in records:  #run them in a for loop, in a form of tuple, put all these values one by one, every time it runs, it adds a new row, one by one, 
    row_cells=table.add_row().cells 
    row_cells[0].text=str(qty)
    row_cells[1].text=id
    row_cells[2].text=desc
document.add_page_break()  #page break means it will got to the next page
document.save('NLPdemo.docx')    
    
    
    
```
